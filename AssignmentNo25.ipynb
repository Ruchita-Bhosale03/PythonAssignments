Square brackets []: When a list comprehension is enclosed in square brackets, it creates a list object. The resulting expression will evaluate to a list containing the elements generated by the comprehension.

Parentheses (): When a list comprehension is enclosed in parentheses, it creates a generator object. The resulting expression will evaluate to a generator, which is an iterable that generates the elements on-the-fly as they are needed, rather than creating the entire list in memory.

Generators are a type of iterator in Python.
Generators are created using generator functions, which use the yield keyword.
Generators provide lazy evaluation and generate values on-demand.
Generators maintain their internal state between iterations.
Generators offer a concise and readable way to create iterators.

Answer3-->
Presence of the yield keyword indicates a generator function.
Generator functions return a generator object instead of executing immediately.

Answer4-->
The purpose of a yield statement is to create a generator object that can produce a sequence of values one at a time, allowing the function to save its state and resume execution from where it left off.

Answer5-->
Map calls require a function and an iterable, returning a map object.
List comprehensions use an expression and an iterable, producing a new list.
Map calls are more flexible as they can handle complex transformations using any function.
List comprehensions are more concise and intuitive for simple transformations or filtering tasks.
List comprehensions are often faster and more efficient than map calls in many cases.
